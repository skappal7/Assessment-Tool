```python
data = pd.read_csv('/content/Analytics Maturity Data.csv')
```

```python
from pycaret.classification import *
s = setup(data, target = 'Level', session_id = 123)
```

```python
best = compare_models(sort = 'AUC', n_select = 15)
compare_model_results = pull()
```

```python
# creating a predict function to be passed into gradio UI
def predict(model, AHT , NTT, Sentiment, Complaints, Repeats):
  
    df = pd.DataFrame.from_dict({'AHT': [AHT], 'NTT': [NTT], 
                                 'Sentiment': [Sentiment], 'Complaints': [Complaints], 'Repeats': [Repeats]})
    
    model_index = list(compare_model_results['Model']).index(model)
    model = best[model_index]
    pred = predict_model(model, df, raw_score=True)
    return {'Nacent_Level': pred['Score_Nacent_Level'][0].astype('float64'), 
            'Low_Maturity': pred['Score_Low_Maturity'][0].astype('float64'),
            'Mid_Maturity' : pred['Score_Mid_Maturity'][0].astype('float64'),
            'High_Maturity' : pred['Score_High_Maturity'][0].astype('float64')}
```

```python
import gradio as gr
```

```python
model = gr.inputs.Dropdown(list(compare_model_results['Model']), label="Model")
AHT = gr.inputs.Slider(minimum=1, maximum=2000, default=data['AHT'].mean(), label = 'AHT')
NTT = gr.inputs.Slider(minimum=1, maximum=500, default=data['NTT'].mean(), label = 'NTT')
Sentiment = gr.inputs.Slider(minimum=-5, maximum=10, default=data['Sentiment'].mean(), label = 'Sentiment')
Complaints = gr.inputs.Slider(minimum=1, maximum=10, default=data['Complaints'].mean(), label = 'Complaints')
Repeats = gr.inputs.Slider(minimum=1, maximum=10, default=data['Repeats'].mean(), label = 'Repeats')

gr.Interface(predict, [model,AHT , NTT, Sentiment, Complaints, Repeats], "label", live=True).launch()
```
